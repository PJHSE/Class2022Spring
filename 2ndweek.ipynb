### Importing libraries

```python
import pandas as pd                     #install, import, built-in
```

### Loading csv file

```python
df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")     #csv 파일을 넣으면 read하는 것. csv? comma separated value? 이걸 액셀 파일로 불러오면 아래처럼. 
df                            #데이터 프레임 변수, 액셀타입
```

```python
print(df.size); print(df.ndim); print(df.shape)               #데이터프레임의 사아즈, 데이터 프레임의 차원, 데이터 프레임의 행 렬. 
```

### Constructing a DataFrame using dict

```python
# Some input data - a small sample of the iris dataset                #dict를 이용해서 csv파일 만들기.
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)                              #dict로 만든 데이터프레임. 
df 
```

### Constructing a DataFrame using pd.Series

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])             #데이터프레임 만드는 또 다른 방법인데, 칼럼 별로 만들어버리기. 시리즈 함수.    
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df                                      #마지막에 딕트 써서 데이터 프레임 형성. 
```

### Changing column names

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']        
df                                     #칼럼 이름 바꿔버리기. 
```

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']        #행, 렬로 액세스 하는 방법. : == 처음부터 끝까지. 
```

```python
# Access one column (short-hand)
df['sepal_length']              #그냥 열의 이름만 적어도 나온다.
```

```python
# Access one column (shortest-hand)
df.sepal_length                  # 또 하나의 방법.
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]
```

```python
# Access two rows
df.loc[[3, 0], :]
```

```python
# Access a range of rows
df.loc[2:5, :]
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]          #특정 정보에 액세스
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4]
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column               #axis = 0, axis = 1 (행, 렬)
df
```

### Adding/Removing new rows

```python
df.loc[5, :] = [1, 2, 3, 4, 'Korea']
df
```

```python
df = df.drop(5, axis=0) # axis=0: row
df
```
